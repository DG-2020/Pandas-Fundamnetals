## 1:- Getting Started

```python
## Importing library
import pandas as pd
```

```python
## Importing Data
sample_data = pd.read_csv('C:/Users/deepe/sample.csv', sep = ',', header = 0)
print(sample_data)
```

```python
## Data Insights ($ sign in has identified by pandas as strings)
#sample_data.groupby('gender')['Salary'].mean()
```

```python
## To make it work we have to write a function which remove the $ sing and create a new column to capture the desired output
def cleanup_salary(row):
    salary = row['Salary'].replace('$', '')
    salary = float(salary) # We are casting a value from string to a float
    return salary

sample_data['Clean_Salary'] = sample_data.apply(cleanup_salary, axis = 1) 
# clean_salary is the result of salary column in the original dataframe, after cleanup salary fucntion has been applied

print(sample_data.groupby('gender')['Clean_Salary'].mean())
```

```python
def female_salary(row):
    if row['gender'] == 'Female':
        return row ['Clean_Salary']

sample_data['Female_Salary_Avg'] = sample_data.apply(female_salary, axis = 1)
print(sample_data)
```

```python
def male_salary(row):
    if row['gender'] == 'Male':
        return row['Clean_Salary']

sample_data['Male_Salary_Avg'] = sample_data.apply(male_salary, axis = 1)
print(sample_data)
```

```python
print(sample_data.groupby('Job Title') ['Clean_Salary', 'Male_Salary_Avg', 'Female_Salary_Avg'].mean())
```

# 3.0:- Exercise Challenge
    Refence to 'exercise_course.csv', we have two problems to answer.
    This exercjise has 6 columns:
        1. username
        2. gender
        3. request time
        4. country
        5. url
        6. bytes
    Problem 1: Find total sum of BYTES per URL
    Problem 2: Find total bytes per domain, split by MALE or FEMALE
    Problem 3: Sum of bytes by male/female and group by URL and COUNTRY

```python
import pandas as pd

ec_df = pd.read_csv('C:\\Users\\deepe\\exercise_course.csv', sep = ',', header = 0)
ec_df.shape
```

```python
ec_df.head(5)
```

```python
ec_df.tail()
```

```python
ec_df.info()
```

```python
ec_df.describe()
```

### Solution 1:

```python
# GROUPBY URL and Sum Bytes
ecdf2 = ec_df.groupby(['url', 'country'])['bytes'].sum().reset_index()

# Sort Values
ecdf2 = ecdf2.sort_values('bytes', ascending = 'False')
print(ecdf2)
```

### Solution 2:

```python
# Define function to calculate total female bytes
def female_bytes(row):
    if row['gender'] == 'Female':
        return row ['bytes']
    
# Define function to calculate total male bytes
def male_bytes(row):
    if row['gender'] == 'Male':
        return row ['bytes']
```

```python
# Create new columns, populated by the functions
ec_df['female_bytes'] = ec_df.apply(female_bytes, axis = 1)

ec_df['male_bytes'] = ec_df.apply(male_bytes, axis = 1)
```

```python
# GROUP BY URL and sum male_bytes and female_bytes
ecdf3 = ec_df.groupby('url')['male_bytes', 'female_bytes'].sum()

print(ecdf3)
```

### Solution 3:

```python
ecdf4 = ec_df.groupby(['country', 'url'])['male_bytes', 'female_bytes'].sum()
print(ecdf4)
```
